```python
from keras.models import load_model
model = load_model("/content/drive/MyDrive/COVID/weights/dense_NET.h5")
```

```python
%matplotlib inline
# get sample image to test.
import numpy as np
import cv2
import matplotlib.pyplot as plt

img_covid = '/content/drive/MyDrive/COVID/test/covid/COVID-3258.png'
img_normal = '/content/drive/MyDrive/COVID/test/normal/Normal-4503.png'


img_array = cv2.imread(img_covid)
plt.imshow(img_array)
#plt.imshow(image)
plt.show()
new_img_array = cv2.resize(img_array, (128, 128))
dt = []
dt.append(new_img_array)
X = np.array(dt)
X = X/255
val = model.predict(X)

if val > 0.5:
    print("COVID")
else:
    print("NORMAL")
```
